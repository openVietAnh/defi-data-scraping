```python
import csv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.dates as md
```

```python
fig, ax = plt.subplots(figsize=(20, 12), dpi=80)

with open("WBTC-usd-price.csv") as csvfile:
    time_range, price = [], []
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None)
    for item in reader:
        time_range.append(int(item[0]) if not False else int(item[0]) / 1000)
        price.append(float(item[1]))

    secs = md.epoch2num(time_range)
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.title("Wrapped bitcoin price")
    ax.ticklabel_format(style='plain', axis='y')
    ax.plot_date(secs, price, "b-", label="WBTC")
    date_fmt = '%d-%m-%y'
    date_formatter = md.DateFormatter(date_fmt)
    ax.xaxis.set_major_formatter(date_formatter)
    fig.autofmt_xdate()
legend = ax.legend(loc="upper left", shadow=True, fontsize='x-large')
plt.show()
```
