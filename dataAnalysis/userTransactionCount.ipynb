```python
import csv
from collections import Counter
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

transactionCount = {}
with open("../csvData/allTransactionType.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        try:
            transactionCount[item[2]] += 1
        except KeyError:
            transactionCount[item[2]] = 1
dct = Counter(transactionCount.values())
x, y = [], []
for i in range(max(dct.keys())):
    x.append(i)
    try:
        y.append(dct[i])
    except KeyError:
        y.append(dct[0])
```
