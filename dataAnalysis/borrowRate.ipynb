```python
import csv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.dates as md
```

```python
files = [
    ("../csvData/stableBorrowRate/DAI-6m-Stable borrow rate-v2.csv", "b-"),
    ("../csvData/variableBorrowRate/DAI-6m-Variable borrow rate-v2.csv", "r-"),
]
fig, ax = plt.subplots(figsize=(20, 12), dpi=80)
rate = [[], []]
for index, f in enumerate(files):
    with open(f[0]) as csvfile:
        time_range = []
        reader = csv.reader(csvfile, delimiter=',')
        next(reader, None)
        for item in reader:
            time_range.append(int(item[0]) / 1000)
            rate[index].append(float(item[2]))

        secs = md.epoch2num(time_range)
        plt.xlabel('Date')
        plt.ylabel('Borrow rate (%)')
        plt.title("DAI Borrow rate")
        ax.plot_date(secs, rate[index], f[1], label=f[0].split("/")[2])
        date_fmt = '%d-%m-%y'

        date_formatter = md.DateFormatter(date_fmt)
        ax.xaxis.set_major_formatter(date_formatter)
        fig.autofmt_xdate()
ax.fill_between(secs, rate[0], rate[1], color='blue', alpha=0.3)
ax.fill_between(secs, rate[1], 0, color='red', alpha=0.3)
legend = ax.legend(loc='center left', shadow=True, fontsize='x-large')
plt.show()
```
