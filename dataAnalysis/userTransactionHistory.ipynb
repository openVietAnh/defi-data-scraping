```python
import csv

user = "0x67e5ab5b573f01030a1076dda7a79ef0f31a1bfa"

transactions = []
with open("../csvData/allTransactionType.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            transactions.append(item)

borrow, deposit, liquidationCall, rebalance, redeem, repay, swap, collateral = {}, {}, {}, {}, {}, {}, {}, {}

with open("../csvData/borrow.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[1] == user:
            borrow[item[0].split(":")[2]] = (item[4], item[3])

with open("../csvData/deposit.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[1] == user:
            if item[0].split(":")[2] == "0x1319eababf5556112706171d9e4b939265171e386628ed210d613bc3a467efa8":
                print("Found")
            deposit[item[0].split(":")[2]] = (item[4], item[3])

with open("../csvData/repay.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            repay[item[0].split(":")[2]] = (item[3], item[5], item[4])

with open("../csvData/usageAsCollateral.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            collateral[item[0].split(":")[2]] = (item[3], item[4], item[5])

with open("../csvData/redeemUnderlying.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            redeem[item[0].split(":")[2]] = (item[3], item[4], item[5])

with open("../csvData/swap.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            swap[item[0].split(":")[2]] = (item[3], item[4], item[5])

with open("../csvData/liquidationCall.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            liquidationCall[item[0].split(":")[2]] = (item[4], item[3], item[6], item[5], item[7])

with open("../csvData/rebalanceStableBorrowRate.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        if item[2] == user:
            rebalance[item[0].split(":")[2]] = (item[3], item[4], item[5])
```

```python
for item in transactions:
    if item[1] == "borrow":
        info = borrow[item[0].split(":")[2]]
        print("Borrowed", float(info[0]) / (10 ** 8), info[1])
    elif item[1] == "deposit":
        info = deposit[item[0].split(":")[2]]
        print("Deposited", float(info[0]) / (10 ** 8), info[1])
    elif item[1] == "repay":
        info = repay[item[0].split(":")[2]]
        print(info[0], "repaid", float(info[1]) / (10 ** 8), info[2], "for user")
    elif item[1] == "redeemUnderlying":
        info = redeem[item[0].split(":")[2]]
        print(info[0], "redeem", float(info[2]) / (10 ** 8), info[1], "for user")
    elif item[1] == "liquidationCall":
        info = liquidationCall[item[0].split(":")[2]]
        print(info[-1], "repay", float(info[2]) / (10 ** 8), info[3], "for user, liquidation", info[0], info[1])
    elif item[1] == "swap":
        info = swap[item[0].split(":")[2]]
        print("Switched", info[0], "borrow mode from", info[1], "to", info[2])
    elif item[1] == "usageAsCollateral":
        info = collateral[item[0].split(":")[2]]
        print("Disable" if info[1] == "True" else "Enable", info[0], "as collateral")
    else:
        info = rebalance[item[0].split(":")[2]]
        print("Change", info[0], "borrow rate from", info[1], "to", info[2])
```
