```python
import csv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.dates as md
```

```python
files = [
    ("../csvData/utilizationRate/DAI-Utilisation rate-v2.csv", "b-"),
    ("../csvData/utilizationRate/USDC-Utilisation rate-v2.csv", "r-"),
    ("../csvData/utilizationRate/USDT-Utilisation rate-v2.csv", "g-"),
    ("../csvData/utilizationRate/WBTC-Utilisation rate-v2.csv", "k-"),
    ("../csvData/utilizationRate/WETH-Utilisation rate-v2.csv", "y-")
]
```

```python
fig, ax = plt.subplots(figsize=(20, 12), dpi=80)
for f in files:
    with open(f[0]) as csvfile:
        time_range, rate = [], []
        reader = csv.reader(csvfile, delimiter=',')
        next(reader, None)
        for item in reader:
            time_range.append(int(item[0]) / 1000 )
            rate.append(float(item[2]))

        secs = md.epoch2num(time_range)
        plt.xlabel('Date')
        plt.ylabel('Utilization rate (%)')
        plt.title("Utilization rate")
        ax.plot_date(secs, rate, f[1], label=f[0].split("/")[3].split("-")[0])
        date_fmt = '%d-%m-%y'

        date_formatter = md.DateFormatter(date_fmt)
        ax.xaxis.set_major_formatter(date_formatter)
        fig.autofmt_xdate()
legend = ax.legend(loc='center left', shadow=True, fontsize='x-large')
plt.show()
```
