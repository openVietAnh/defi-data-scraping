```python
import csv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.dates as md
```

```python
fig, ax = plt.subplots(figsize=(20, 12), dpi=80)

with open("../csvData/WBTC-usd-price.csv") as csvfile:
    time_range, price = [], []
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None)
    for item in reader:
        time_range.append(int(item[0]))
        price.append(float(item[1]))

    secs = md.epoch2num(time_range)
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.title("WBTC price")
    ax.ticklabel_format(style='plain', axis='y')
    ax.plot_date(secs, price, "b-", label="WBTC")
    date_fmt = '%d-%m-%y'
    date_formatter = md.DateFormatter(date_fmt)
    ax.xaxis.set_major_formatter(date_formatter)
    fig.autofmt_xdate()
legend = ax.legend(loc='center left', shadow=True, fontsize='x-large')
plt.show()
```

```python
files = [
    ("../csvData/WETH-usd-price.csv", "b-"),
    ("../csvData/WBTC-usd-price.csv", "r-"),
]
fig, ax = plt.subplots(figsize=(20, 12), dpi=80)
for f in files:
    with open(f[0]) as csvfile:
        time_range, price = [], []
        reader = csv.reader(csvfile, delimiter=',')
        next(reader, None)
        for item in reader:
            time_range.append(int(item[0]))
            price.append(float(item[1]))
        secs = md.epoch2num(time_range)
        plt.xlabel('Date')
        plt.ylabel('Price ($)')
        plt.title("WETH and WBTC price comparison")
        ax.plot_date(secs, price, f[1], label=f[0].split("/")[2].split("-")[0])
        date_fmt = '%d-%m-%y'

        date_formatter = md.DateFormatter(date_fmt)
        ax.xaxis.set_major_formatter(date_formatter)
        fig.autofmt_xdate()
legend = ax.legend(loc='center left', shadow=True, fontsize='x-large')
plt.show()
```
