```python
import csv
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.dates as md
```

```python
users = set()
time, count = [], []
with open("../csvData/allTransaction.csv", "r") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    next(reader, None)
    for item in reader:
        time.append(int(item[-1]))
        users.add(item[2])
        count.append(len(users))
```

```python
fig, ax = plt.subplots(figsize=(20, 12), dpi=80)
secs = md.epoch2num(time)
plt.xlabel("Date")
plt.ylabel("Unique address count")
plt.title("Total number of address")
ax.ticklabel_format(style='plain', axis='y')
ax.plot_date(secs, count, "b-", label="Address count")
date_fmt = '%d-%m-%y'
date_formatter = md.DateFormatter(date_fmt)
ax.xaxis.set_major_formatter(date_formatter)
fig.autofmt_xdate()
legend = ax.legend(loc="upper left", shadow=True, fontsize='x-large')
plt.show()
fig.savefig('../exportedImages/totalUniqueAddressCount.png', bbox_inches='tight', transparent=False)
```
