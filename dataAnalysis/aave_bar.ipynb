```python
import csv
fees = []
with open("../aaveDaily/aave_fee.csv", "r") as csv_file:
    reader = csv.reader(csv_file, delimiter=",")
    next(reader, None)
    for item in reader:
        fees.append(float(item[2]))
```

```python
fees
```

```python
import numpy as np
import matplotlib.pyplot as plt
  
fig = plt.figure(figsize = (20, 10))
 
plt.bar([i for i in range(len(fees))], fees, color ='maroon',width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
